```python
# imports
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D

# set the plot color
color_base = (0.5, 0.2, 0.8)
color_finetuned = (0.627, 0.992, 0.945)
```

```python
# data import
ds = pd.read_csv('Final_DS_Eval.csv')
bk = pd.read_csv('Final_BK_Eval.csv')
db = pd.read_csv('Final_db_Eval.csv')
bd = pd.read_csv('Df_Big_D.csv')
ml = pd.read_csv('Final_ML_Eval.csv')

all_df = [ds, bk, db, bd, ml]
```

```python
# first look
ds.head()
```

GPT wins

```python
# get the wins for each model in each dataset

gpt_wins_base_ds = (ds['Result'] == 1).sum()
gpt_wins_finetuned_ds = (ds['Result'] == 2).sum()

gpt_wins_base_bk = (bk['Result'] == 1).sum()
gpt_wins_finetuned_bk = (bk['Result'] == 2).sum()

gpt_wins_base_db = (db['Result'] == 1).sum()
gpt_wins_finetuned_db = (db['Result'] == 2).sum()

gpt_wins_base_bd = (bd['Result'] == 1).sum()
gpt_wins_finetuned_bd = (bd['Result'] == 2).sum()

gpt_wins_base_ml = (ml['Result'] == 1).sum()
gpt_wins_finetuned_ml = (ml['Result'] == 2).sum()

# sum up all wins
total_gpt_wins_base = gpt_wins_base_ds + gpt_wins_base_bk + gpt_wins_base_db + gpt_wins_base_bd + gpt_wins_base_ml
total_gpt_wins_finetuned = gpt_wins_finetuned_ds + gpt_wins_finetuned_bk + gpt_wins_finetuned_db + gpt_wins_finetuned_bd + gpt_wins_finetuned_ml

# print the total wins
print("Total GPT Wins (Base):", total_gpt_wins_base)
print("Total GPT Wins (Fine-tuned):", total_gpt_wins_finetuned)
```

```python
# plot the gpt wins
fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(0, gpt_wins_base_ds, color=color_base, label='Llama-base')
ax.bar(1, gpt_wins_finetuned_ds, color=color_finetuned, label='LLama-finetuned')

ax.bar(3, gpt_wins_base_bk, color=color_base)
ax.bar(4, gpt_wins_finetuned_bk, color=color_finetuned)

ax.bar(6, gpt_wins_base_db, color=color_base)
ax.bar(7, gpt_wins_finetuned_db, color=color_finetuned)

ax.bar(9, gpt_wins_base_bd, color=color_base)
ax.bar(10, gpt_wins_finetuned_bd, color=color_finetuned)

ax.bar(12, gpt_wins_base_ml, color=color_base)
ax.bar(13, gpt_wins_finetuned_ml, color=color_finetuned)

ax.set_xticks([0.5, 3.5, 6.5, 9.5, 12.5])
ax.set_xticklabels(['Data Science\nFundamentals', 'Kommunikations-\nund Betriebssysteme', 'Datenbanken', 'Big Data', 'Machine\nLearning'])
ax.set_yticks([5, 10, 15])
ax.set_ylabel('Anzahl der Siege')

ax.set_title('Ergebnisse des Finetunings:\nGPT-4 Richter')
#fig.suptitle(f'Gesamtsiege\nLlama-base: {total_gpt_wins_base} | Llama-finetuned: {total_gpt_wins_finetuned}', y=0)

ax.legend()

plt.savefig('gpt_wins.png', dpi=500, bbox_inches='tight')
```

Similarity wins

```python
# get the wins for each model in each dataset
similarity_wins_base_ds = (ds['similarity_winner'] == 1).sum()
similarity_wins_finetuned_ds = (ds['similarity_winner'] == 2).sum()

similarity_wins_base_bk = (bk['similarity_winner'] == 1).sum()
similarity_wins_finetuned_bk = (bk['similarity_winner'] == 2).sum()

similarity_wins_base_db = (db['similarity_winner'] == 1).sum()
similarity_wins_finetuned_db = (db['similarity_winner'] == 2).sum()

similarity_wins_base_bd = (bd['similarity_winner'] == 1).sum()
similarity_wins_finetuned_bd = (bd['similarity_winner'] == 2).sum()

similarity_wins_base_ml = (ml['similarity_winner'] == 1).sum()
similarity_wins_finetuned_ml = (ml['similarity_winner'] == 2).sum()

# sum up all wins
total_similarity_wins_base = similarity_wins_base_ds + similarity_wins_base_bk + similarity_wins_base_db + similarity_wins_base_bd + similarity_wins_base_ml
total_similarity_wins_finetuned = similarity_wins_finetuned_ds + similarity_wins_finetuned_bk + similarity_wins_finetuned_db + similarity_wins_finetuned_bd + similarity_wins_finetuned_ml

# print the total wins
print("Total Similarity Wins (Base):", total_similarity_wins_base)
print("Total Similarity Wins (Fine-tuned):", total_similarity_wins_finetuned)
```

```python
# plot the similarity wins
fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(0, similarity_wins_base_ds, color=color_base, label='Llama-base')
ax.bar(1, similarity_wins_finetuned_ds, color=color_finetuned, label='LLama-finetuned')

ax.bar(3, similarity_wins_base_bk, color=color_base)
ax.bar(4, similarity_wins_finetuned_bk, color=color_finetuned)

ax.bar(6, similarity_wins_base_db, color=color_base)
ax.bar(7, similarity_wins_finetuned_db, color=color_finetuned)

ax.bar(9, similarity_wins_base_bd, color=color_base)
ax.bar(10, similarity_wins_finetuned_bd, color=color_finetuned)

ax.bar(12, similarity_wins_base_ml, color=color_base)
ax.bar(13, similarity_wins_finetuned_ml, color=color_finetuned)

ax.set_xticks([0.5, 3.5, 6.5, 9.5, 12.5])
ax.set_xticklabels(['Data Science\nFundamentals', 'Kommunikations-\nund Betriebssysteme', 'Datenbanken', 'Big Data', 'Machine\nLearning'])
ax.set_yticks([5, 10, 15])
ax.set_ylabel('Anzahl der Siege')

ax.set_title('Ergebnisse des Finetunings:\nKosinusähnlichkeit zur Standardantwort')
#fig.suptitle(f'Gesamtsiege\nLlama-base: {total_similarity_wins_base} | Llama-finetuned: {total_similarity_wins_finetuned}', y=0)

ax.legend()

plt.savefig('similarity_wins.png', dpi=500, bbox_inches='tight')
```

Median similarity

```python
# get the median similarity for each model in each dataset
median_similarity_base_ds = ds['similarity_1'].median()
median_similarity_finetuned_ds = ds['similarity_2'].median()

median_similarity_base_bk = bk['similarity_1'].median()
median_similarity_finetuned_bk = bk['similarity_2'].median()

median_similarity_base_db = db['similarity_1'].median()
median_similarity_finetuned_db = db['similarity_2'].median()

median_similarity_base_bd = bd['similarity_1'].median()
median_similarity_finetuned_bd = bd['similarity_2'].median()

median_similarity_base_ml = ml['similarity_1'].median()
median_similarity_finetuned_ml = ml['similarity_2'].median()
```

Average similarity

```python
# get the average similarity for each model in each dataset
avg_similarity_base_ds = ds['similarity_1'].mean()
avg_similarity_finetuned_ds = ds['similarity_2'].mean()

avg_similarity_base_bk = bk['similarity_1'].mean()
avg_similarity_finetuned_bk = bk['similarity_2'].mean()

avg_similarity_base_db = db['similarity_1'].mean()
avg_similarity_finetuned_db = db['similarity_2'].mean()

avg_similarity_base_bd = bd['similarity_1'].mean()
avg_similarity_finetuned_bd = bd['similarity_2'].mean()

avg_similarity_base_ml = ml['similarity_1'].mean()
avg_similarity_finetuned_ml = ml['similarity_2'].mean()

total_avg_similarity_base = (avg_similarity_base_ds + avg_similarity_base_bk + avg_similarity_base_db + avg_similarity_base_bd + avg_similarity_base_ml) / 5
total_avg_similarity_finetuned = (avg_similarity_finetuned_ds + avg_similarity_finetuned_bk + avg_similarity_finetuned_db + avg_similarity_finetuned_bd + avg_similarity_finetuned_ml) / 5
```

```python
# plot the average similarity

fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(0, avg_similarity_base_ds, color=color_base, label='Llama-base')
ax.bar(1, avg_similarity_finetuned_ds, color=color_finetuned, label='LLama-finetuned')

ax.bar(3, avg_similarity_base_bk, color=color_base)
ax.bar(4, avg_similarity_finetuned_bk, color=color_finetuned)

ax.bar(6, avg_similarity_base_db, color=color_base)
ax.bar(7, avg_similarity_finetuned_db, color=color_finetuned)

ax.bar(9, avg_similarity_base_bd, color=color_base)
ax.bar(10, avg_similarity_finetuned_bd, color=color_finetuned)

ax.bar(12, avg_similarity_base_ml, color=color_base)
ax.bar(13, avg_similarity_finetuned_ml, color=color_finetuned)

ax.set_xticks([0.5, 3.5, 6.5, 9.5, 12.5])
ax.set_xticklabels(['Data Science\nFundamentals', 'Kommunikations-\nund Betriebssysteme', 'Datenbanken', 'Big Data', 'Machine\nLearning'])
ax.set_ylabel('Durchschnittliche Kosinusähnlichkeit')

ax.set_title('Ergebnisse des Finetunings:\nKosinusähnlichkeit zur Standardantwort')
fig.suptitle(f'Durchschnittliche Kosinusähnlichkeit\nLlama-base: {round(total_avg_similarity_base, 2)} | Llama-finetuned: {round(total_avg_similarity_finetuned, 2)}', y=0)

ax.legend()

plt.savefig('avg_similarity.png', dpi=500, bbox_inches='tight')
```
